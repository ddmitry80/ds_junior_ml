```python
import numpy as np
import pandas as pd
```

```python
# загружаем данные
phones = pd.read_csv("data/1.8_phones.csv")
```

```python
phones.head(17)
```

```python
# Заменяем поле 'os' числовым аналогом
phones['os_d']=phones['os']=='Android'
```

```python
phones
```

```python
# импортируем модуль, отвечающий за деревья решений
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
```

```python
# выгружаем признаки и целевые значения в отдельные переменные
X = phones[["disk", "year", "os_d"]]
y = phones["price"]

# создаем классификатор
cl = DecisionTreeClassifier().fit(X, y)
```

```python
# выведем информацию для интерпретации построенной модели
print(export_text(cl))
```

```python
# проведем классификацию
cl.predict(X[7:8])[0], y[7]
```
